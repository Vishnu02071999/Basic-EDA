```python
# usually represented in the form of Nan null or None value
'''ml algos fail if the dataset contins missing values however algos like K-nearest and naive bayes support data with missing values
ml models can become biased if this is not handled properly'''
'''Types of MIssing values-->MCAR, MAR, MNAR
----MAR — Missing At Random

✅ Meaning:

The missingness is related to observed data, but not to the missing value itself.

Once you account for other variables, the missingness is random.
# Income values missing more often for younger people
# (missingness depends on "Age", which is observed)


-----MNAR — Missing Not At Random

🚫 Meaning:

The missingness depends on the missing value itself (i.e., the reason for being missing is directly tied to the value that’s missing).
 People with very high income choose not to disclose their income.
# Missingness depends on the actual income value (unobserved).


------MCAR — Missing Completely At Random

✅ Meaning:

The missingness of a value is completely independent of any observed or unobserved data.

In other words, data is missing for no particular reason.

📊 Example:

# Example: Sensor data missing due to random transmission failure
# Missing values occur purely by chance
```

Implementing KNN imputer

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# Sample DataFrame
data = {
    "Height": [170, 168, 172, 167],
    "Weight": [65, np.nan, 70, 60],
    "Age":    [25, 26, 24, 23]
}

df = pd.DataFrame(data, index=[1,2,3,4])

print("Original DataFrame:")
print(df)

# Create KNN Imputer (k = 2 neighbors)
imputer = KNNImputer(n_neighbors=2)

imputed_data=imputer.fit_transform(df) #imputing missing values..output will be stored in form of array

# Fit and transform the data
df_imputed = pd.DataFrame(imputed_data, columns=df.columns,index=[1,2,3,4])   # converted into dataframe

print("\nImputed DataFrame:")
print(df_imputed)
```
